```python
df2= sns.load_dataset('titanic')  # this data default available
```

```python
df2.head()
```

```python
df2.iplot(x = 'age',y= 'fare', mode= 'markers')  # able to see how 'age' and 'fare' bahaving, (random varition fare wrt age, no relation)
```

```python
df2.iplot(x = 'age',y= 'pclass', mode= 'markers')  
```

```python
 #how 'age' behaves with pclass, classes are equally distrubuted wrt age, 
```

```python
df2.iplot(x = 'fare',y= 'pclass', mode= 'markers')
```

```python
# for pclass one more fare compared to other pclass
```
